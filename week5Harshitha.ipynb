```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("SalesPerformance.xlsx")

# Create total wait time
df["Total_Wait"] = df["Incoming call wait"] + df["During call wait"]
```

```python
call_purpose_counts = df["Call purpose"].value_counts()

fig, ax = plt.subplots(figsize=(8,5))

ax.bar(call_purpose_counts.index,
       call_purpose_counts.values)

ax.set_title("Call Volume by Purpose")
ax.set_xlabel("Call Purpose")
ax.set_ylabel("Number of Calls")

ax.tick_params(axis='x', rotation=45)

plt.show()
```

```python
grouped_data = [
    df[df["Call purpose"] == purpose]["Total_Wait"]
    for purpose in df["Call purpose"].unique()
]

fig, ax = plt.subplots(figsize=(8,5))

ax.boxplot(grouped_data, tick_labels=df["Call purpose"].unique())

ax.set_title("Total Wait Time Distribution by Call Purpose")
ax.set_xlabel("Call Purpose")
ax.set_ylabel("Total Wait Time")

plt.xticks(rotation=45)
plt.show()
```

```python
grouped_data_branch = [
    df[df["Branch"] == branch]["Total_Wait"]
    for branch in df["Branch"].unique()
]

fig, ax = plt.subplots(figsize=(8,5))

ax.boxplot(grouped_data_branch, tick_labels=df["Branch"].unique())

ax.set_title("Distribution of Total Wait Time by Branch")
ax.set_xlabel("Branch")
ax.set_ylabel("Total Wait Time")

plt.show()
```
